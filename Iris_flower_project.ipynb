importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

importing dataset

```python
url=" https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width","class"]
iris_data = pd.read_csv(url, names=column_names)
```

```python
iris_data.head(10)
```

```python
iris_data.describe()
```

Data analysis using pairplot from seaborn libraries.

```python
sns.pairplot(iris_data, hue="class")
plt.show()
```

Seprateing the variable "class" as dependent variable 'y' and independent variable 'x'

```python
x= iris_data.drop("class",axis=1)
x
```

```python
y = iris_data["class"]
y
```

spliting the data-set into training and testing data_set using sklearn function 'train_test_split' 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=42 )
```

using fit function in knn librerary to fit our training or testin   x_train, y_train data set

```python
knn=KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
```

now we will check the accuricy of our model by evoluating it.

```python
y_pred = knn.predict(x_test)
```

compairing the pridicted value by actule value.

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

model evaluation

```python
print(classification_report(y_test, y_pred))
```

Thats all we successfully created our knn = "KNeighborsClassifier" model using the data set and we have got the accuracy score of 1 whish is 100% accurate.
